```python
# simple test function
def simple_eighths(num_notes, pitch, tempo):
    beat = 0.0
    seq = music_pb2.NoteSequence()
    while num_notes/2 != beat:
        seq.notes.add(pitch = pitch, start_time=beat, end_time=beat + 0.5, velocity=80)
        beat += 0.5
    seq.total_time = beat
    seq.tempos.add(qpm=tempo)
    return seq
        
test_eighths = simple_eighths(8, 60, 200)
note_seq.play_sequence(test_eighths)
```
